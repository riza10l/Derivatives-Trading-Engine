## 1. Avellaneda-Stoikov Baseline
Closed-form market making model: $\delta = \sigma^2\tau/2 + (1/\gamma)\ln(1+\gamma/\kappa)$

```python
env = MarketMakingEnv(steps=2000, volatility=0.3)
pnl_baseline = run_baseline(env)
print(f"Baseline final PnL: {pnl_baseline[-1]:.2f}")
print(f"Trades: {env.mm.trades}")
```

## 2. RL Agent
Training PPO/SAC dengan stable-baselines3.

```python
try:
    import stable_baselines3 as sb3
    agent = RLAgent(env, algo='PPO')
    agent.train(env, total_timesteps=100_000)
    pnl_rl = run_rl(env, agent)
    print(f"RL final PnL: {pnl_rl[-1]:.2f}")
except ImportError:
    print("Install stable-baselines3: pip install stable-baselines3 gymnasium")
```

## 3. Perbandingan
Hasil benchmark RL vs Avellaneda-Stoikov.

```python
try:
    results = benchmark(pnl_baseline, pnl_rl)
    for k, v in results.items():
        print(f"  {k}: {v}")
except NameError:
    print("Train RL agent first")
```
